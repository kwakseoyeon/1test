```python
def simple_model(input_text):
    # Í∏çÏ†ïÏ†ÅÏù∏ ÌïµÏã¨Ïñ¥ Î∞è n-gramsÏôÄ Í∞ÄÏ§ëÏπò
    positive_keywords = {
        "ÏµúÍ≥†": 3,
        "Ï∂îÏ≤ú": 3,
        "ÏßÑÏßú Ï¢ãÎã§": 2.5,
        "ÎåÄÎ∞ï": 3,
        "Í∑πÏ∞¨": 3,
        "ÏïÑÎäë": 2,
        "Ï¢ãÏïòÏäµÎãàÎã§": 2,
        "ÎßõÏßë": 3,
        "ÏßÑÎ¶¨": 3,
        "Ìï´Ìîå": 2.5,
        "ÎßõÏûà": 2.5,
        "Ïò§Í≥† Ïã∂Îã§": 2.5,
        "ÎÑàÎ¨¥ ÏπúÏ†àÌï¥ÏÑú": 2,
        "Í∏∞Î∂Ñ Ï¢ãÍ≤å": 1.5,
        "Î∂ÑÏúÑÍ∏∞ÎèÑ Ï¢ãÍ≥†": 1.5,
        "ÏßÑÏßú „Ñπ„Öá": 2,
        "ÎßõÏûàÏñ¥Ïöî": 2,
        "Ï¢ãÏùÄ": 1,
        "ÏôîÎäîÎç∞": 1,
        "Ï†ïÎßê": 1.5,
        "Ï¢ãÏïÑÏöî": 1
    }

    # Î∂ÄÏ†ïÏ†ÅÏù∏ ÌïµÏã¨Ïñ¥ Î∞è n-gramsÏôÄ Í∞ÄÏ§ëÏπò
    negative_keywords = {
        "Ïã§Îßù": 3,
        "ÏïÑÏâΩÎã§": 2.5,
        "Í∑∏ÎÉ• Í∑∏Îûò": 2.5,
        "Ïã§ÎßùÏä§Îü¨ÏõÄ": 3,
        "ÏùåÏãùÏù¥ ÎÑàÎ¨¥": 2,
        "ÎÑàÎ¨¥ ÏßúÏÑú": 1.5,
        "Í∑∏ÎÉ• Í∑∏Îû¨Ïñ¥Ïöî": 2,
        "Í∏∞ÎåÄ Ïù¥ÌïòÏòÄÏäµÎãàÎã§": 2,
        "Îä¶Í≤å ÎÇòÏò§Í≥†": 1.5,
        "Î≥ÑÎ°úÏòÄÏñ¥Ïöî": 2,
        "Ïù¥Î†áÍ≤å": 1,
        "ÎÑàÎ¨¥": 1
    }

    # ÌÇ§ÏõåÎìúÎ•º Í∏∏Ïù¥Ïóê Îî∞Îùº Ï†ïÎ†¨
    sorted_positive_keywords = sorted(positive_keywords.keys(), key=len, reverse=True)
    sorted_negative_keywords = sorted(negative_keywords.keys(), key=len, reverse=True)

    positive_score = 0
    negative_score = 0

    # Í∏çÏ†ï ÌÇ§ÏõåÎìú Ï†êÏàò Í≥ÑÏÇ∞
    for key in sorted_positive_keywords:
        if key in input_text:
            positive_score += positive_keywords[key]
            input_text = input_text.replace(key, "")  # Ï§ëÎ≥µ Ï†êÏàò Î∞©ÏßÄ

    # Î∂ÄÏ†ï ÌÇ§ÏõåÎìú Ï†êÏàò Í≥ÑÏÇ∞
    for key in sorted_negative_keywords:
        if key in input_text:
            negative_score += negative_keywords[key]
            input_text = input_text.replace(key, "")  # Ï§ëÎ≥µ Ï†êÏàò Î∞©ÏßÄ

    # Í≤∞Í≥º Î∞òÌôò
    if positive_score > negative_score:
        return 1
    else:
        return 0
```

```python
import urllib.request

urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarr_train.txt",
    filename="tarr_train.txt",
)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

def classify_text(input_text):
    output = simple_model(input_text)
    return output

# ÌååÏùºÏùÑ DataFrameÏúºÎ°ú Î°úÎìú
df = pd.read_csv('tarr_train.txt', delimiter='\t')

actual_labels = []
predicted_labels = []

for index, row in df.iterrows():
    comment = row['comment']
    actual_label = row['label']
    predicted_label = classify_text(comment)

    actual_labels.append(actual_label)
    predicted_labels.append(predicted_label)

    print("comment : ", comment)
    print("actual class : ", actual_label)
    print("predicted class : ", predicted_label)
    print("---------------")

# Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞
accuracy = accuracy_score(actual_labels, predicted_labels)
print(f"Accuracy: {accuracy*100:.2f}%")

print(set(actual_labels))
print(set(predicted_labels))

# Confusion matrix Í≥ÑÏÇ∞
cm = confusion_matrix(actual_labels, predicted_labels, labels=[1, 0])
# Confusion matrix ÌëúÌòÑ
print("\nConfusion Matrix:")
print("         Predicted:")
print("         Í∏çÏ†ï    Î∂ÄÏ†ï")
print("Actual")
print("Í∏çÏ†ï      {:<5}  {:<5}".format(cm[0][0], cm[0][1]))
print("Î∂ÄÏ†ï      {:<5}  {:<5}".format(cm[1][0], cm[1][1]))
```
